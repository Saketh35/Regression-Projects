```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('APY.csv')
```

```python
df.head()
```

```python
df.nunique()
```

```python
df['State']=df['State'].replace({'THE DADRA AND NAGAR HAVELI':'Dadra and Nagar Haveli'})
```

```python
le=LabelEncoder()
```

```python
df['Crop']=le.fit_transform(df['Crop'])
df['District ']=le.fit_transform(df['District '])
```

```python
df=df.join(pd.get_dummies(df['State']))
df=df.join(pd.get_dummies(df['Season']))
```

```python
df.dropna(inplace=True)
```

```python
df['Area ']=np.log(df['Area ']+1)
```

```python
sns.distplot(df['Area '])
```

```python
s=StandardScaler()
```

```python
df.iloc[:,5:7]=s.fit_transform(df.iloc[:,5:7])
```

```python
x=df.drop(columns=['State','Season','Yield'])
y=df['Yield']
```

```python
from sklearn.model_selection import train_test_split
```

```python
a,b,c,d=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
model=RandomForestRegressor()
model.fit(a,c)
model.score(b,d)
```

```python
model.score(a,c)
```

```python
model=DecisionTreeRegressor()
model.fit(a,c)
model.score(a,c)
```
